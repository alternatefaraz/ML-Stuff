```python
#Remove columns with NA values more than 30%
dfBugs = dfImportedBugs.loc[:,dfImportedBugs.isna().sum() < 90]
dfBugs.columns
```

```python
#Droping columns that are not required
dfBugs.drop(["Work Item Type", "Activated By", "Authorized As", "State Change Date", "Submit Date", "Submitter", "Team Project", "Column1", "Watermark"], axis=1, inplace=True)
```

```python
#Selecting only bugs that are closed
dfBugs =  dfBugs[dfBugs.State == "Closed"]
dfBugs.count()
```

```python
#See total developers and their bugs fixed count
dfBugs.groupby('Resolved By').agg(['count']).sum(level=0)
```

```python
#See how many bugs are not actual bugs so they can be removed
dfBugs.groupby('Resolved Reason').agg(['count'])
```

```python
#Removing bugs from the list which are not bugs
replaceString = {'Duplicate': np.NaN, 'Cannot Reproduce': np.NaN, 'Obselete': np.NaN}
dfBugs.loc[:, "Resolved Reason"] = dfBugs["Resolved Reason"].replace(replaceString)
dfBugs.loc[:, "Resolved Reason"] = dfBugs.dropna(subset=["Resolved Reason"])
dfBugs.count()
```

```python
#Calculate total fixing time
dfBugs["TotalFixTime"] = (dfBugs["Resolved Date"] - dfBugs["Activated Date"])/np.timedelta64(1,'h')
print(dfBugs["TotalFixTime"].isna().sum())
modeValue = dfBugs["TotalFixTime"].mode()
print(modeValue)
dfBugs["TotalFixTime"] = dfBugs["TotalFixTime"].fillna(float(modeValue))
print(dfBugs["TotalFixTime"].isna().sum()) 
```

```python
dfBugs.groupby(['Node Name', 'Area Path']).agg(['count'])
```

```python
#Use customer column to distinguish between internal or external bug
dfBugs["IsCustomer"] = dfBugs["Customer"].apply(lambda x: 1 if x == "POMS" else 0)
dfBugs.loc[:, ["Customer", "IsCustomer"]].head(5)  
```

```python
#Checking severity
dfBugs.groupby(['Severity']).agg(['count'])
```

```python
#Extract area for area path
extractionDict = {}
minorList = []
re.split("\\\\", "\POMSnet")
def ExtractArea(str):
    extracted = re.split("\\\\", str)
    lenExtracted = len(extracted)
    if(lenExtracted == 0):
        return np.NaN
    elif(lenExtracted == 1 | lenExtracted == 2):
        return np.NaN
    elif(lenExtracted >= 3):
        if(extracted[2] in extractionDict):
            if(lenExtracted == 4):
                minorList = extractionDict[extracted[2]]
                if(extracted[3] not in minorList):
                    minorList.append(extracted[3])
                extractionDict.update({extracted[2]: minorList})
        elif(extracted[2] not in extractionDict):
            extractionDict.update({extracted[2]: []})
        return extracted[2]

dfBugs["MajorArea"] = dfBugs["Area Path"].apply(lambda x: ExtractArea(x.lower()))   
#dfBugs["MajorArea"]
print(extractionDict)
```

```python
dfBugs.groupby(['MajorArea']).agg(['count'])
```

```python
print(dfBugs['MajorArea'].isna().sum())
dfBugs[dfBugs['MajorArea'].isna()]
```

```python
def TokenizeWork(title):
    lowerTitle = str(title).lower()    
    for dicKey, dicValues in extractionDict.items():
        if(re.search(dicKey, lowerTitle)):
            return dicKey
        for val in dicValues:
            if(re.search(val, lowerTitle)): 
                return dicKey

#dfBugs["MajorArea"] = 
#.apply(lambda x: TokenizeWork(x["Title"]))

filledValues = dfBugs[dfBugs["MajorArea"].isnull()].apply(lambda x: TokenizeWork(x["Title"]), axis=1)
dfBugs["MajorArea"] = dfBugs["MajorArea"].fillna(value=filledValues) 
```

```python
dfBugs.dropna(subset=['Resolved By'], inplace=True)
```

```python
dfBugsOneHot = dfBugs.copy()
dfBugsOneHot = dfBugsOneHot.loc[:, ["MajorArea", "IsCustomer", "TotalFixTime", "Priority"]]
dfBugsOneHot = pd.get_dummies(dfBugsOneHot, columns=['MajorArea'], prefix = ['MajorArea'])
dfBugsOneHot.head(5)

dfBugsOneHot.reindex()
X = dfBugsOneHot
y = dfBugs.loc[:, ["Resolved By"]]
print(X.info())
print(y.info())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)

#knn = KNeighborsClassifier(n_neighbors=8)
#knn.fit(X_train, y_train.values.ravel())

#predictions = knn.predict(X_test)

#print(metrics.accuracy_score(y_test, predictions))
#print(metrics.confusion_matrix(y_test, predictions))
#print(metrics.classification_report(y_test, predictions))


#clf = DecisionTreeClassifier(max_depth=100)
#clf = clf.fit(X_train, y_train.values.ravel())
#y_pred = clf.predict(X_test)
#print("Decision Tree Accuracy:",metrics.accuracy_score(y_test, y_pred))

scv = SVC(kernel='linear',C=1.0)
scv.fit(X_train, y_train.values.ravel()) 
y_pred = scv.predict(X_test)
print("SVC Accuracy:",metrics.accuracy_score(y_test, y_pred))

model = GaussianNB()
model.fit(X_train, y_train.values.ravel())
y_pred = model.predict(X_test)
print("Naye Bayes Accuracy:",metrics.accuracy_score(y_test, y_pred))



```
